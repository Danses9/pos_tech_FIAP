## Operadores aritiméticos
### Soma (+), subtração(-), multiplicação(*), divisão(/), divisão inteira(//), módulo(%) e 
### exponenciação(**)

```python
Abacate = 10
Ameixa = 3

print("Soma:", Abacate + Ameixa)
print("Multiplicação:", Abacate * Ameixa)
print("Subtração:", Abacate - Ameixa)
print("Divisão:", Abacate / Ameixa)
print("Divisão inteira:", Abacate // Ameixa)
```

## Operadores relacionais
#### Comparam valores e retornam resultados booleanos (**VERDADEIRO ou FALSO**)
#### Utilizando: ==, !=, >, <, >= e <= 


```python
x = 8
y = 5

print(x < y)
print(y > x)
```

## Operadores lógicos:
#### Permitem combinar condições com **AND, OR e NOT**

```python
idade = 26
tem_habilitacao = True

print(idade >= 18 and tem_habilitacao)
```

## Operadores de atribuição
#### Atualizam variáveis de forma compacta com **=, +=, -=**, etc

```python
contador = 21
print(contador)

contador += 5
print(contador)
```

## Estruturas condicionais 
#### **IF** verifica uma condição e se for verdadeira, executa o bloco associado;
#### **ELIF** permite testar condições alternativas;
#### **ELSE** executa um bloco padrão quando nenhuma condição anterior é satisfeita 

```python
nota = 8

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

## Estruturas Aninhadas
#### Você pode combinar condições e criar estruturas aninhadas para resolver situações mais complexas

```python
idade = 16
habilitacao = True

if idade >= 18:
    if habilitacao:
        print("Pode dirigir")
    else:
        print("Precisa tirar carteira de habilitação :( ")
```

## Operadores lógicos em condicionais
#### Combinar condições com **AND, OR** para testes mais robustos

```python
idade = 22 
experiencia = 3

if idade >= 23 and experiencia >= 2:
    print("Elegivel para a vaga")
else: 
    print("Não é elegivel para a vaga")
```
