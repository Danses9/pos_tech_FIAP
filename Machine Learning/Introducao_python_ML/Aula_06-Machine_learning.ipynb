#### NumPy - Operações numéricas com Arrays Objetivo: Aprender a criar e manipular arrays, que são a base para operações numéricas eficiêntes.

### 1. Importação e Criação de arrays:

```python
import numpy as np

# Criando um array a partir de uma lista
lista = [1, 2, 3, 4, 5]
array_1 = np.array(lista)
print("Array 1", array_1)

# Criando um array 2D (matriz)
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("Matriz: \n", matriz)
```

### 2. Operações básicas com Arrays

```python
# Soma, multiplicação e operações element-wise
array_2 = np.array([10, 20, 30, 40, 50,])
soma = array_1 + array_2
multiplicacao = array_1 * array_2
print("Soma:",soma)
print("Multiplicação:",multiplicacao)
```

## Pandas - Manipulação e análise de dados
### Objetivo: Utilizar o Pandar para carregar, explorar e manipular conjuntos de dados de forma eficiente.

#### 1. Importação e criação do Data Frame

```python
import pandas as pd

# Criando um data frame a partir de um dicionário
dados = {
    'Nome': ['Ana', 'Raimunda', 'Vicente'],
    'Idade': [25, 30, 45],
    'Salario': [50000, 70000, 88000]
}
df = pd.DataFrame(dados)
print("DataFrame: \n", df)
```

#### Manipulação e exploração de dados

```python
# Selecionando uma coluna
print("Coluna 'Nome':\n, df['Nome]")

# Filtrando linhas com base em uma condição
df_filtrado = df[df['Salario'] > 6000]
print("Funcionário com salário > 6000:\n", df_filtrado)

# Estatísticas descritivas
print("Estatísticas:\n", df.describe())
```

#### Seaborn - Visualizações avançadas com o objetivo de criar gráficos estatísticos aprimoraas e insigths aprofundados

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Configuração padrão para gráficos mais bonitos
sns.set(style="whitegrid")
```

#### Criação de gráficos com Seaborn

```python
print(df.columns)
```

```python
# Gráfico de dispersão
sns.scatterplot(x='Idade', y= 'Salario', data=df, hue='Nome', s=100)
plt.title("Scatter Plot: Salário vs Idade")
plt.show()

```

#### Boxplot

```python
# Boxplot para visualizar a distribuição do salário
sns.boxplot(y= 'Salario', data=df)
plt.title("Distribuição de salário")
plt.show()
```

### Histograma

```python
# Gráfico de distribuição
sns.histplot(df['Salario'], kde=True)
plt.title("Histograma do salário")
plt.show()
```
