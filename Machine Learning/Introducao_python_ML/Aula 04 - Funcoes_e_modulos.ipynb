### Simples exemplos de funções

```python
def somar(a, b):
    return a + b 

# Usando a função
resultado = somar(3, 4)
print("Resultado da somatória: ", resultado)
```

### Exemplo de função com docstring
### São documentações que explica o que a função faz

```python
def calcular_media(numeros):
    """ Calcular a média de uma lista de números.

    Parâmetros:
        números (list): Lista contendo números (int ou float)

    Retorna:
        float: A média dos valores na lista.
    """
    if not numeros:
        return 0
    return sum(numeros) / len(numeros)

# Usando a função
valores = [10, 20, 30, 40, 50]
print("Média: ", calcular_media(valores))
```

#### É importante quando houver muito uso de funções, gerar um novo documento onde você possa concentrar todas as funções geradas(ex: utilidades) e um outro documento onde você possa usa-la (main)
