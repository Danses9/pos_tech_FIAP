## Programação orientada a objetos é um modo de pensar e organizar o código

### 1. Conceitos básicos: Classes, objetos, métodos e construtores

```python
class Carro:                                        # Criando uma classe
    def __init__(self, modelo, cor, ano):           # Criando um construtor
        self.modelo = modelo                        # Self representa o objeto criado
        self.cor = cor                              # Parâmetros do construtor: modelo, cor, ano
        self.ano = ano                              # Atributos do objeto: modelo, cor, ano
    
    def acelerar(self, velocidade):                 # Método da classe e parâmetro velocidade
        # Método que simula a aceleração do carro
        return f"0 {self.modelo} está acelerando {velocidade} km/h." # Devolve um valor

# Criando um objeto da classe Carro
meu_carro = Carro("EX40", "Bright Dusk", 2026)                   # Criando o objeto

print(meu_carro.acelerar(140))                                         # Chamando o método
```

### 2. Princípios ativos de POO

#### Herança
#### Evitando repetições e gerando hierarquia

```python
class Veiculo:
    def __init__(self, marca, ano):
        self.marca = marca
        self.ano = ano

    def descricao(self):
        return f"Veículo da marca {self.marca}, ano {self.ano}"
    
class Carro(Veiculo):
    def __init__ (self, marca, ano, modelo):
        # Utiliza-se herança para iniciar marca e ano
        super().__init__(marca, ano)
        self.modelo = modelo

meu_carro = Carro("Volvo", 2026, "EX40")
print(meu_carro.descricao())
```

### Polimorfismo
#### Diferentes classes usando o mesmo método com comportamentos específicos

```python
class Animal:
    def emitir_som(self):
        pass

class Gato(Animal):
    def emitir_som(self):
        return "Miau"

class Cachorro(Animal):
    def emitir_som(self):
        return "Au Au"
    
class Passaro(Animal):
    def emitir_som(self):
        return "Bem-te-vi"
    
animais = [Gato(), Cachorro(), Passaro()]

for animal in animais:
    print(animal.emitir_som()) # Todos os sons
```

### Encapsulamento
#### Esconder os detalhes do objeto, expondo apenas a classe

```python
class ContaBancaria:
    def __init__(self, saldo_inicial):
         self.__saldo = saldo_inicial # Atributo privado

    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor
    
    def obter_saldo(self):
        return self.__saldo
    
# Exemplo de uso
conta = ContaBancaria (1100)
conta.depositar(290)
print("Saldo:", conta.obter_saldo())
```

#### O saldo é privado e só pode ser acessado por meio de get_saldo()
