### Numpy

```python
# Importando a biblioteca Numpy
import numpy as np
```

```python
# Criando um array (lista) usando a biblioteca Numpy
arr_list = np.array([1, 2, 3, 4, 5, 6, 7, 8])

print(arr_list)
```

```python
# Criando uma matriz
arr_zeros = np.zeros((4,6)) # 4 linhas e 6 colunas preenchidas com zeros
print(arr_zeros)
```

```python
arr_ones = np.ones((3,4))
print(arr_ones)
```

```python
# Criando uma matriz randomica
arr_random = np.random.rand(3, 4)
print(arr_random)
```

```python
print(arr_random.shape)
```

```python
array_random_reshape = arr_random.reshape((4, 3))
print(array_random_reshape)
```

```python
# Criando matrizes 2D
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
arr3 = np.array([[9, 10], [11, 12]])
```

```python
arr1
```

```python
# Concatenando as matrizes horizontalmente (axis=1)
arr4 = np.concatenate((arr1, arr2, arr3), axis=1)
arr4
```

```python
# Separando os arrays em 2 partes iguais
arr4_split = np.split(arr4, 2)
print(arr4_split)
```

```python
# Realizando trasposição da matriz arr4
arr4_transpose = np.transpose(arr4)
print(arr4_transpose)
```

```python
# Revertendo o valor 
arr4_revertido = arr4_transpose.T
arr4_revertido
```

```python
# Somando os arrays
arr_a = np.array([1, 7, 28])
arr_b = np.array([1, 5, 8])
```

```python
arr_a_b = np.add(arr_a, arr_b)
arr_a_b
```

```python
arr_sub_a_b = np.subtract(arr_a, arr_b)
arr_sub_a_b
```
