```python
# Criando uma função para categorizar os tomates com base na média do peso
def categorizar_tomate_media(media):
    if media >= 40 and media <= 70:
        return "Tomate médio"
    elif media < 40:
        return "Tomate pequeno"
    else:
        return "Tomatão"
```

```python
# Aplicando a função de categorização à coluna "Average" do dataframe do csv e criando uma nova coluna "categoria_tomate"
df_csv["categoria_tomate"] = df_csv["Average"].apply(categorizar_tomate_media)
```

```python
df_csv.head(3)
```

```python
# Categorizando os tomates com base na média do peso e visualizando as estatísticas descritivas agrupadas por categoria
df_csv.groupby("categoria_tomate").describe()
```

```python
# Criando um filtro selecionando os tomates com a média menor que 40
filtro = df_csv["Average"] < 40
df_csv.loc[filtro] # O .loc no Pandas é usado para acessar linhas e colunas usando rótulos e filtros lógicos.
```

```python
# Importando a biblioteca
import matplotlib.pyplot as plt
```

```python
# Avaliar as colunas
df_csv.info()
```

```python
# Convertendo a coluna "Date" para o tipo datetime
df_csv["Date"] = pd.to_datetime(df_csv["Date"])
```

```python
# Verificando a transformação da coluna "Date"
df_csv.info()
```

```python
# Criando um gráfico de linha para visualizar a média do peso dos tomates ao longo do tempo
plt.plot(df_csv["Date"], df_csv["Average"])

plt.xlabel("Data") # Rótulo do eixo x
plt.ylabel("Média e peso dos tomates") # Rótulo do eixo y
plt.title("Média do peso dos tomates ao longo do tempo") # Título do gráfico

plt.show() # Exibe o gráfico
```

```python
# Trazendo o dataframe do excel: Chess
df_excel.head()
```

```python
# Criando um gráfico de dispersão para visualizar a relação entre as classificações das peças de xadrez
plt.scatter(df_excel["black_rating"], df_excel["white_rating"])
plt.xlabel("Black")
plt.ylabel("White")
plt.title("Relação entre as classificações das cores das peças de xadrez")

plt.show()
```

```python
# Gráfico de barras para visualizar a média do peso dos tomares por categoria
plt.bar(df_csv["categoria_tomate"], df_csv["Average"], color="#B32A24")
plt.xlabel("Categoria de Tomates")
plt.ylabel("Média do peso dos tomates")
plt.title("Média de tomates por categorias")

plt.show()
```

```python
# Gráfico de pizza para visualizar a média de partidas por status de vitória
df_excel.groupby("victory_status")["turns"].mean().plot(kind="pie", y='turns', autopct="%1.0f%%")
plt.title("Média de partidas por status de vitória")

plt.show()
```
